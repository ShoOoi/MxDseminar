### Step.4 Show the reult images

```
import cv2
import matplotlib.pyplot as plt
import os.path

# 座標軸の削除処理
fig,ax = plt.subplots()
ax.tick_params(labelbottom="off",bottom="off")
ax.tick_params(labelleft="off",left="off")
ax.set_xticklabels([]) 
ax.axis('off')

path = './images/gen_imgs/'
files =[path + 'img_0.png', path + 'img_200.png', path + 'img_400.png', path + 'img_600.png', path + 'img_800.png', path + 'img_1000.png', 
        path + 'img_1200.png', path + 'img_1400.png', path + 'img_1600.png', path + 'img_1800.png', path + 'img_2000.png']

plt.figure(figsize=(30, 24))
for i, file in enumerate(files):
  img = cv2.imread(file)
  plt.subplot(3,5,i+1)
  show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
  plt.imshow(show_img)
```
